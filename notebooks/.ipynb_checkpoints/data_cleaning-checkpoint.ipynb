```python
n_america = ['Canada', 'Costa Rica', 'Mexico', 'United States']
europe = ['Austria', 'Belgium', 'Switzerland', 'Czech Republic', 'Germany', 'Denmark', 'Spain', 'Estonia', 'Finland', 'France', 'United Kingdom', 'Greece', 'Hungary', 'Ireland', 'Iceland', 'Italy', 'Lithuania', 'Luxembourg', 'Latvia', 'Netherlands', 'Norway', 'Poland', 'Portugal', 'Slovenia', 'Sweden']
asia = ['Chile', 'Isreal', 'Japan', 'Turkey']
oceania = ['Australia', 'New Zealand']

def get_continents(continent):
    if continent in n_america:
        continent = 'N_America'
        return continent
    
    elif continent in europe:
        continent = 'Europe'
        return continent
    
    elif continent in asia:
        continent = 'Asia'
        return continent
    
    else:
        continent = 'Oceania'
        return continent
```

```python
df['Continent'] = df.Country.apply(get_continents)
```

```python
coffee_prices = pd.read_csv('/Users/joelporcaro/Metis/project2/coffee_regression/raw_data/coffee_prices.csv', skiprows=3)
coffee_prices.head()
```

```python
coffee_prices = coffee_prices.rename({'Calendar years': 'Country'}, axis=1)
coffee_prices.head()
```

```python
coffee_prices = coffee_prices.drop(coffee_prices.index[0])
coffee_prices
```

```python
coffee_prices.columns
```

```python
prices_years = pd.Series(coffee_prices.columns[1:])
prices_years
```

```python
prices_df = pd.DataFrame()

for country_count in range(28):
    country = coffee_prices.iloc[country_count][0]
    current_country = pd.Series(country)
    current_country = current_country.repeat(28).reset_index().loc[:,0]
    country_plus_years = pd.concat([current_country, prices_years], axis=1)
    current_country_values = coffee_prices.loc[country_count+1].values[1:]
    
    values_series = pd.concat([country_plus_years, pd.Series(current_country_values)], axis=1)
#     values_series
    prices_df = pd.concat([prices_df, values_series])
    prices_df = prices_df.reset_index(drop=True)
    
prices_df
```

```python
prices_df
```

```python
prices_df.columns = ['Country', 'Year', 'Price']
```

```python
prices_df['Year'] = pd.to_numeric(prices_df['Year'])
```

```python
main_df = pd.merge(df, prices_df, on=['Country', 'Year'], how='left')
main_df
```
