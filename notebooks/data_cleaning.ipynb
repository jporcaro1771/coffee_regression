```python
unemployment_df = unemployment_df.drop(columns=['Unnamed: 0'], axis=1)
unemployment_df.head()
```

```python
hours_worked_df = pd.read_csv("hours_worked_df.csv")
hours_worked_df.head()
```

```python
income_age_merge = pd.merge(income_df, age_df)
income_age_edu_merge = pd.merge(income_age_merge, edu_df)
income_age_edu_unemployment_merge = pd.merge(income_age_edu_merge, unemployment_df)
df = pd.merge(income_age_edu_unemployment_merge, hours_worked_df)
```
